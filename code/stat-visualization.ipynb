```python
import pandas as pd
import matplotlib.pyplot as plt

# Provided data
data = {
    'Dataset': ['ChatGPT 2024', 'PubMed 2010-2014', 'PubMed 2015-2019', 'PubMed 2020-2024'],
    "Alzheimer's_Nodes": [519, 742, 774, 817],
    "Alzheimer's_Edges": [1194, 861, 940, 958],
    'Cancer_Nodes': [559, 755, 828, 817],
    'Cancer_Edges': [1050, 803, 977, 1030],
    'Depression_Nodes': [577, 802, 775, 790],
    'Depression_Edges': [1108, 878, 826, 809]
}

df = pd.DataFrame(data)  # Make sure to replace this line with your actual DataFrame

# Calculate ratios
df["Alzheimer's_Ratio"] = df["Alzheimer's_Nodes"] / df["Alzheimer's_Edges"]
df['Cancer_Ratio'] = df['Cancer_Nodes'] / df['Cancer_Edges']
df['Depression_Ratio'] = df['Depression_Nodes'] / df['Depression_Edges']

# Define colors for ChatGPT and Publication ratios
chatgpt_color = 'royalblue'
pub_color = 'lightcoral'  # You can choose a different color

# Plot the ratios for all three diseases
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.bar(df['Dataset'], df["Alzheimer's_Ratio"], color=[chatgpt_color if 'ChatGPT' in dataset else pub_color for dataset in df['Dataset']])
plt.title("Alzheimer's Nodes to Edges Ratio")
plt.xlabel('Dataset')
plt.ylabel('Ratio (Nodes to Edges)')

plt.subplot(3, 1, 2)
plt.bar(df['Dataset'], df['Cancer_Ratio'], color=[chatgpt_color if 'ChatGPT' in dataset else pub_color for dataset in df['Dataset']])
plt.title('Cancer Nodes to Edges Ratio')
plt.xlabel('Dataset')
plt.ylabel('Ratio (Nodes to Edges)')

plt.subplot(3, 1, 3)
plt.bar(df['Dataset'], df['Depression_Ratio'], color=[chatgpt_color if 'ChatGPT' in dataset else pub_color for dataset in df['Dataset']])
plt.title('Depression Nodes to Edges Ratio')
plt.xlabel('Dataset')
plt.ylabel('Ratio (Nodes to Edges)')

plt.tight_layout()
# plt.show()

# Save the plot to a file (e.g., PNG)
plt.savefig('../figures/premise_test1.png')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the data
data = {
    'Disease': ['Alzheimer\'s', 'Depression', 'Cancer'],
    'Correct': [76, 50, 82],
    'Misclassification': [24, 50, 18],
    'PubMed_Correct': [100, 100, 100],
    'PubMed_Misclassification': [0, 0, 0],
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width
bar_width = 0.35

# Bar positions
r1 = range(len(df['Disease']))
r2 = [x + bar_width for x in r1]

# Plot bars
plt.bar(r1, df['Correct'], color='blue', width=bar_width, edgecolor='grey', label='Correct')
plt.bar(r1, df['Misclassification'], color='lightblue', width=bar_width, edgecolor='grey', label='Misclassification', bottom=df['Correct'])

# Set labels and title
plt.xlabel('Disease')
plt.ylabel('Percentage')
plt.title('Multi-Mode - Correct vs Misclassification of ChatGPT')
plt.xticks([r + bar_width/2 for r in range(len(df['Disease']))], df['Disease'])

# Move the legend to the right outside the plot
plt.legend(loc='center left', bbox_to_anchor=(0.58, .90))

# Save the plot to a file
plt.savefig('../figures/chatgpt-multi-mode.png')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Dataset': ['2010-2014', '2015-2019', '2020-2024'],
    'Alzheimer\'s': [78, 80, 76],
    'Depression': [80, 50, 50],
    'Cancer': [88, 86, 82]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width
bar_width = 0.25

# Bar positions
r1 = range(len(df['Dataset']))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Plot bars
plt.bar(r1, df['Alzheimer\'s'], width=bar_width, label='Alzheimer\'s')
plt.bar(r2, df['Depression'], width=bar_width, label='Depression')
plt.bar(r3, df['Cancer'], width=bar_width, label='Cancer')

# Set labels and title
plt.xlabel('Dataset')
plt.ylabel('Correct Classification (%)')
plt.title('Mult-Mode: ChatGPT Correct Classification for Different Diseases and Periods')
plt.xticks([r + bar_width for r in range(len(df['Dataset']))], df['Dataset'])

# Place the legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Save to a file
plt.savefig('../figures/year-factor.png', bbox_inches='tight')

# Display the plot
plt.show()
```
