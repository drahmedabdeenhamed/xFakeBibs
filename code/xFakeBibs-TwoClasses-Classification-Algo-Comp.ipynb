```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your DataFrame)
data = {
    'Classifier': ['xFakeScience', 'Naive Bayes', 'Linear SVM', 'Classical SVM', 'Logistic Regression'],
    'CAN 2010 - 2014': [94, 46, 52, 50, 50],
    'CAN 2015 - 2019': [92, 43, 51, 47, 47],
    'CAN 2020 - 2024': [91, 43, 51, 47, 46],
    'ALZ 2010 - 2014': [89, 45, 48, 41, 48],
    'ALZ 2015 - 2019': [90, 44, 48, 40, 46],
    'ALZ 2020 - 2024': [89, 47, 50, 41, 49],
    'DEP 2010 - 2014': [80, 51, 38, 42, 42],
    'DEP 2015 - 2019': [80, 50, 40, 44, 45],
    'DEP 2020 - 2024': [80, 52, 39, 44, 44]
}
df = pd.DataFrame(data)
df.set_index('Classifier', inplace=True)

# Plot stacked bar plot with nested grouping
plt.figure(figsize=(12, 6))

# Plot CAN F1 scores
plt.subplot(1, 3, 1)
df_can = df.filter(like='CAN')
df_can.plot(kind='bar', stacked=True, ax=plt.gca(), color=['skyblue', 'dodgerblue', 'darkblue'])
plt.title('F1 Scores for Cancer')
plt.ylabel('F1 Score (%)')
plt.xlabel('Classifier')
plt.xticks(rotation=45)
plt.legend(title='Period')

# Plot ALZ F1 scores
plt.subplot(1, 3, 2)
df_alz = df.filter(like='ALZ')
df_alz.plot(kind='bar', stacked=True, ax=plt.gca(), color=['lightgreen', 'limegreen', 'forestgreen'])
plt.title("F1 Scores for Alzheimer's")
plt.ylabel('F1 Score (%)')
plt.xlabel('Classifier')
plt.xticks(rotation=45)
plt.legend(title='Period')

# Plot DEP F1 scores
plt.subplot(1, 3, 3)
df_dep = df.filter(like='DEP')
df_dep.plot(kind='bar', stacked=True, ax=plt.gca(), color=['lightcoral', 'indianred', 'darkred'])
plt.title('F1 Scores for Depression')
plt.ylabel('F1 Score (%)')
plt.xlabel('Classifier')
plt.xticks(rotation=45)
plt.legend(title='Period')

plt.tight_layout()

# Save the figure to a file
plt.savefig('../figures/f1_disease_periods.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {
    'Classifier': ['xFakeScience', 'Naive Bayes', 'Linear SVM', 'Classical SVM', 'Logistic Regression'],
    'CAN 2020 - 2024': [91, 43, 51, 47, 46],
    'ALZ 2020 - 2024': [89, 47, 50, 41, 49],
    'DEP 2020 - 2024': [80, 52, 39, 44, 44]
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Set the 'Classifier' column as the index
df.set_index('Classifier', inplace=True)

# Transpose the DataFrame for easier plotting
df = df.T

# Plotting
plt.figure(figsize=(10, 6))
df.plot(kind='bar', stacked=True)
plt.title('F1 Scores by xFakeSci Classifier')
plt.xlabel('Disease')
plt.ylabel('F1 Score')
plt.xticks(rotation=45)
plt.legend(title='Classifier', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.savefig('../figures/f1_diseases_year19to24.png', dpi=300)

plt.show()
```
